```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
'''
Columns are:
1. Average Rating (the arithmetic mean of all individual quality ratings of this professor)
2. Average Difficulty (the arithmetic mean of all individual difficulty ratings of this professor)
3. Number of ratings (simply the total number of ratings these averages are based on)
4. Received a “pepper”? (Boolean - was this professor judged as “hot” by the students?)
5. The proportion of students that said they would take the class again
6. The number of ratings coming from online classes
7. Male gender (Boolean – 1: determined with high confidence that professor is male)
8. Female (Boolean – 1: determined with high confidence that professor is female)
'''
numDat = pd.read_csv('rmpCapstoneNum.csv', header=None)
numDat.columns = ['Average Rating', 'Average Difficulty', 'Number of ratings', 'Received a pepper', 
                       'Proportion of students that said they would take the class again', 
                       'Number of ratings coming from online classes', 'Male Professor', 'Female Professor']
```

```python
# We will only look at professors with 10 or more ratings

# Keep rows where column 'Number of ratings' greater or equal to 10
numDat_filtered = numDat.loc[numDat['Number of ratings'] >= 10]

numDat_filtered
```

```python
# Also controlling for online class as confound... number of ratings - online ratings must be greater or equal to ten.... Then we will have 
# teachers with 10 or more ratings based off of iun person class

numDat_filtered_for_online = numDat_filtered.loc[(numDat_filtered['Number of ratings'] - numDat_filtered['Number of ratings coming from online classes']) >= 10]

numDat_filtered_for_online
```

```python
### We also control for the gender... We look at those who are confidently male/female

numDat_filtered_for_online_MF = numDat_filtered_for_online[(numDat_filtered_for_online['Male Professor'] != numDat_filtered_for_online['Female Professor'])]

numDat_filtered_for_online_MF
```

```python
#### Below result might not generalize well (for online classes(?))... But it will apply to male and female professors and for in person classes...

import matplotlib.pyplot as plt

# Separate male and female professor data
male_data1 = numDat_filtered_for_online_MF[numDat_filtered_for_online_MF['Male Professor'] == 1]['Average Rating']
female_data1 = numDat_filtered_for_online_MF[numDat_filtered_for_online_MF['Female Professor'] == 1]['Average Rating']

# Create subplots for histograms
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Male professor histogram
axes[0].hist(male_data1, bins=10, alpha=0.7, color='blue', density=True)
axes[0].set_title('Histogram of Average Ratings (Male Professors)', fontsize=14)
axes[0].set_xlabel('Average Rating', fontsize=12)
axes[0].set_ylabel('Density', fontsize=12)

# Female professor histogram
axes[1].hist(female_data1, bins=10, alpha=0.7, color='orange', density=True)
axes[1].set_title('Histogram of Average Ratings (Female Professors)', fontsize=14)
axes[1].set_xlabel('Average Rating', fontsize=12)

# Adjust layout
plt.tight_layout()
plt.show()

```

```python
### We also control for the gender... We look at those who are confidently male/female

numDat_filtered_MF = numDat_filtered[(numDat_filtered['Male Professor'] != numDat_filtered['Female Professor'])]

numDat_filtered_MF

#### Below result might not generalize well (for online classes(?))... But it will apply to male and female professors and for in person classes...

import matplotlib.pyplot as plt

# Separate male and female professor data
male_data2 = numDat_filtered_MF[numDat_filtered_MF['Male Professor'] == 1]['Average Rating']
female_data2 = numDat_filtered_MF[numDat_filtered_MF['Female Professor'] == 1]['Average Rating']

# Create subplots for histograms
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Male professor histogram
axes[0].hist(male_data2, bins=10, alpha=0.7, color='blue', density=True)
axes[0].set_title('Histogram of Average Ratings (Male Professors)', fontsize=14)
axes[0].set_xlabel('Average Rating', fontsize=12)
axes[0].set_ylabel('Density', fontsize=12)

# Female professor histogram
axes[1].hist(female_data2, bins=10, alpha=0.7, color='orange', density=True)
axes[1].set_title('Histogram of Average Ratings (Female Professors)', fontsize=14)
axes[1].set_xlabel('Average Rating', fontsize=12)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
### I test difference in variation of female with online and female without online ... same with male... Right
# right now i am not testing male versus female i am just seeing if online classes leads to significant difference... this will help me decide
# if i should keep or remove them

from scipy.stats import levene

# Test variance for male professors
stat_male, p_male = levene(male_data1, male_data2)
print(f"Male Professors - Levene's Test: Statistic={stat_male}, p-value={p_male}")

# Test variance for female professors
stat_female, p_female = levene(female_data1, female_data2)
print(f"Female Professors - Levene's Test: Statistic={stat_female}, p-value={p_female}")

# Interpret results
if p_male < 0.005:
    print("Significant difference in variance between male_data1 and male_data2.")
else:
    print("No significant difference in variance between male_data1 and male_data2.")

if p_female < 0.005:
    print("Significant difference in variance between female_data1 and female_data2.")
else:
    print("No significant difference in variance between female_data1 and female_data2.")
```

From abopve we can concliude that cobtrolling for online classes does not impact the variance of the two groups significantly

```python
### can we test to see if the distributions are significantly different when controlling for online class???
from scipy.stats import ks_2samp

# Test distribution for male professors
stat_male_dist, p_male_dist = ks_2samp(male_data1, male_data2)
print(f"Male Professors - KS Test: Statistic={stat_male_dist}, p-value={p_male_dist}")

# Test distribution for female professors
stat_female_dist, p_female_dist = ks_2samp(female_data1, female_data2)
print(f"Female Professors - KS Test: Statistic={stat_female_dist}, p-value={p_female_dist}")

# Interpret results
if p_male_dist < 0.005:
    print("Significant difference in distribution between male_data1 and male_data2.")
else:
    print("No significant difference in distribution between male_data1 and male_data2.")

if p_female_dist < 0.005:
    print("Significant difference in distribution between female_data1 and female_data2.")
else:
    print("No significant difference in distribution between female_data1 and female_data2.")

```

From above we also see the distribution isnt significantly impacted when considering online classes...

```python
# Now, we test the variance between male and female groups... we will first use the data controlled for online classes

from scipy.stats import levene

# Test variance between male_data1 and female_data1
stat1, p1 = levene(male_data1, female_data1)
print(f"Male_Data1 vs Female_Data1 - Levene's Test: Statistic={stat1}, p-value={p1}")

# Test variance between male_data2 and female_data2
stat2, p2 = levene(male_data2, female_data2)
print(f"Male_Data2 vs Female_Data2 - Levene's Test: Statistic={stat2}, p-value={p2}")

# Interpret results for male_data1 vs female_data1
if p1 < 0.005:
    print("Significant difference in variance between male_data1 and female_data1.")
else:
    print("No significant difference in variance between male_data1 and female_data1.")

# Interpret results for male_data2 vs female_data2
if p2 < 0.005:
    print("Significant difference in variance between male_data2 and female_data2.")
else:
    print("No significant difference in variance between male_data2 and female_data2.")
```

In both cases there is a significant difference in the variance of ratings between male and female subjects.

```python
import numpy as np

# Calculate mean and standard deviation
mean_male1 = np.mean(male_data1)
mean_female1 = np.mean(female_data1)
std_male1 = np.std(male_data1, ddof=1)
std_female1 = np.std(female_data1, ddof=1)

# Coefficient of variation
cv_male1 = std_male1 / mean_male1
cv_female1 = std_female1 / mean_female1

print(f"CV for Male Professors (Group 1): {cv_male1}")
print(f"CV for Female Professors (Group 1): {cv_female1}")
```

```python
import numpy as np

def bootstrap_variance_diff(data1, data2, n_resamples=1000, sample_size=3000):
    var_diffs = []
    for _ in range(n_resamples):
        sample1 = np.random.choice(data1, size=sample_size, replace=True)
        sample2 = np.random.choice(data2, size=sample_size, replace=True)
        var_diffs.append(np.var(sample1, ddof=1) - np.var(sample2, ddof=1))
    return np.percentile(var_diffs, [2.5, 97.5])  # 95% confidence interval

# Bootstrap variance difference
ci_variance_diff = bootstrap_variance_diff(male_data2, female_data2)
print(f"95% Confidence Interval for Variance Difference: {ci_variance_diff}")
```

Interpretation of the Interval
The fact that the interval is entirely negative indicates:
Male ratings are consistently less variable (smaller variance) than female ratings.
There is 95% confidence that the true difference in variances lies between 
−0.1325 and −0.0058.

```python
# Number of experiments
num_experiments = 100000

p_values = []

for _ in range(num_experiments):
    # Sample len(male_data2) times from male_data2, and len(female_data2) times from female_data2
    male_sample = np.random.choice(male_data2, len(male_data2), replace=True)
    female_sample = np.random.choice(female_data2, len(female_data2), replace=True)
    
    # Run Levene's test
    _, p_value = levene(male_sample, female_sample)
    
    # Collect the p-value
    p_values.append(p_value)
    print(p_value)
# Calculate average p-value
average_p_value = np.mean(p_values)

print(f"Average p-value from Levene's test across {num_experiments} experiments: {average_p_value:.5f}")
```

```python
# Calculate the group mean for male_data2
male_group_mean = np.mean(male_data2)
male_differences = male_data2 - male_group_mean

# Calculate the group mean for female_data2
female_group_mean = np.mean(female_data2)
female_differences = female_data2 - female_group_mean

# Plot the histogram for both datasets
plt.figure(figsize=(8, 6))

# Male differences histogram
plt.hist(
    male_differences, bins=20, edgecolor='black', alpha=0.6, label="Male Differences", color="blue"
)

# Female differences histogram
plt.hist(
    female_differences, bins=20, edgecolor='black', alpha=0.6, label="Female Differences", color="orange"
)

# Add labels, title, and legend
plt.title("Histogram of Differences from Group Mean")
plt.xlabel("Difference from Group Mean")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
from scipy.stats import ks_2samp

# Perform the KS test
stat, p_value = ks_2samp(male_differences, female_differences)

print(f"KS Test Statistic: {stat:.5f}")
print(f"p-value: {p_value:.5f}")

# Interpretation
alpha = 0.005  # Significance level
if p_value < alpha:
    print("The distributions are significantly different.")
else:
    print("The distributions are not significantly different.")
```

```python
from scipy.stats import mannwhitneyu

# Perform the Mann-Whitney U test
stat, p_value = mannwhitneyu(male_differences, female_differences, alternative='two-sided')

print(f"Mann-Whitney U Test Statistic: {stat:.5f}")
print(f"p-value: {p_value:.5f}")

# Interpretation
if p_value < alpha:
    print("The distributions are significantly different.")
else:
    print("The distributions are not significantly different.")
```
